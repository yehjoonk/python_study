# 1. Fibonacci 수열 재귀호출 (Recursive Programming)

```python
def fib_rec(n):
    if n < 3:
        return 1
    else:
        return (fib_rec(n-1)+fib_rec(n-2))
```

```python
print(fib_rec(100))
```

# 2. Fibonacci 수열 동적처리 (Dynamic Programming)

```python
def fib_dp(n):
    table = [0,1]
    for i in range(2, n+1):
        table.append(table[i-1] + table[i-2])
    return (table[n])
```

```python
fib_dp(7)
```

```python
fib_dp(1000)
```
