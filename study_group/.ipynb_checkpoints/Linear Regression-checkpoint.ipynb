## Linear Regression

```python
%matplotlib inline 
import numpy as np
from pylab import *

page_speeds = np.random.normal(3.0,1.0,1000)

#first para = mean, 2nd para = standard dev, 3r para = no. of data points
print(page_speeds)
```

```python
purchase_amount= 100 - (page_speeds - np.random.normal(0, 0.1, 1000)) * 3
print(purchase_amount)
```

```python
scatter(page_speeds, purchase_amount)
```

```python
from scipy import stats

slope, intercept, r_value, p_value, std_error = stats.linregress(page_speeds, purchase_amount)


print(slope)
print(intercept)
print(r_value**2)
print(p_value)
print(std_error)
```

```python
def predict(x):
    return slope * x + intercept

fitLine = predict(page_speeds)

scatter(page_speeds, purchase_amount)
plot(page_speeds, fitLine, c = 'Red')
```
