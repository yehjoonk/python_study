```python
import pandas as pd
```

```python
df = pd.read_excel("/Users/macbook/Desktop/example1.xlsx", skiprows = 2, sheet_name = "Sheet1")
```

```python
df = df.drop([2], axis = 0)
```

```python
df = df.dropna(axis = 1)
```

```python
df = df.drop_duplicates(["ID"])
```

```python
df = df.replace("790101-223457S", "790101-223457")
```

```python
df = df.reset_index()
```

```python
df = df.drop(["index"], axis = 1)
```

```python
df["age"] = "아무값"
```

```python
df["gender"] = "이것도"
```

```python
for i in range(0, 23):
    df["age"][i] = df.values[i,1][0:2]
    
for i in range(0,23):
    df["gender"][i] = df.values[i,1][7]
    
df
```

```python
for i in range(0,23):
    if df.values[i,4] > "21" and df.values[i,3] < "3":
        df.values[i,4] = "adult"
    elif df.values[i,4] <= "21" and df.values[i,3] < "3":
        df.values[i,4] = "adult"
    else:
        df.values[i,4] = "child"
        
df
```

```python
for i in range (0,23):
    if df.values[i,3] == "1" or df.values[i,3] == "3":
        df.values[i,3] = "Male"
    else:
        df.values[i,3] = "Female"

df
```
