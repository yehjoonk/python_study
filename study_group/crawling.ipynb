```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(req.content, "html.parser")
print(soup)
```

```python
soup.find_all("span")
span_list = soup.find("tbody").find_all("span")
span_list[0]
```

```python
span_list2 = []
for span in span_list:
    span_list2.append(span.text)
    
print(span_list2)
```

```python
del span_list2[0]
span_list3 = []
for span in span_list2:
    if len(span)>=4:
        span_list3.append(span)

print(span_list3)
```

```python
f = open("/Users/macbook/desktop/voca.text","w", encoding = "utf-8")
for span in span_list3:
    f.write(span+"\n")
f.close()
```

```python
g = open("/Users/macbook/desktop/voca.csv","w", encoding = "utf-8")
for span in span_list3:
    g.write(span+"\n")
g.close()
```
