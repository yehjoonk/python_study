```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

### Importing and Exploring Dataset

```python
bmw = pd.read_csv("bmw.csv")
bmw.head(10)
```

```python
bmw.isnull().sum()
```

```python
bmw.describe()
```

```python
bmw['age'] = 2021 - bmw.year

bmw.loc[bmw['model'] == ' 3 Series', 'segment'] = 'D'
bmw.loc[bmw['model'] == ' 5 Series', 'segment'] = 'E'
bmw.loc[bmw['model'] == ' 7 Series', 'segment'] = 'F'

bmw.head()
```

```python
bmw.dropna(inplace=True)
bmw.head()
```

```python
sns.pairplot(bmw)
```

```python
sns.countplot(bmw["fuelType"])
```

```python
sns.countplot(bmw["age"])
```

```python
sns.countplot(bmw["segment"])
```
